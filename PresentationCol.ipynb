Color Dictionary

```python
def DiffColors(Annotation):
    Clr_Dict={0:'Lead',1:'Position', 2:'Counterclaim',3:'Rebuttal', 4:'Concluding Statement',5:'Evidence',6:'Claim'}
    try:
        num=Clr_Dict.get_key(Annotation)
    except KeyError:
        print('no such key')
    return num
    
```

Painting with a specific color

```python
def paintS(string, Annotation):
    painted_strin=bg(DiffColors(Annotation))+string+rs.bg
    return painted_strin
    
    
```
